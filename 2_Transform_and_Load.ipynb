```python
#Transforming and Loading Student Load Debt data from excel.
import pandas as pd
from sqlalchemy import create_engine
debt_data = 'extracted_student_loan_data.xlsx'
engine = create_engine('postgresql://<USERNAME>:<PASSWORD>@localhost:5432/mydb')
```

```python
#Read excel. Store in a data frame.
county_debt = pd.read_excel(debt_data,
sheet_name="Student Loan Debt - County",
header=0,
index_col=False,
keep_default_na=True
)

state_debt = pd.read_excel(debt_data,
sheet_name="Student Loan Debt - State, US",
header=0,
index_col=False,
keep_default_na=True
)
type(county_debt)
```

```python
#Test.
county_debt.head()
state_debt.head()
```

```python
#Dropping NAs and irrelevant columns for county data.
county_debt = county_debt.dropna(axis=1,how='all')
county_debt = county_debt.drop(columns=['Nonwhite population share'])

county_debt.head()
```

```python
#Dropping NAs and irrelevant columns for state data.
state_debt = state_debt.dropna(axis=1,how='all')
state_debt = state_debt.drop(columns=['Nonwhite population share'])

state_debt.head()
```

```python
#Changing header level.
headers = county_debt.iloc[0]
headers

#Defining new master data frame.
dfc = pd.DataFrame(county_debt.values[1:], columns=headers)
dfc.head()
```

```python
#Changing header level.
headers = state_debt.iloc[0]
headers

#Defining new master data frame.
dfs = pd.DataFrame(state_debt.values[1:], columns=headers)
dfs.head()
```

```python
#Maintaining original headers by storing in a list. KateColumns is the list of all debt metric categories.
KateColumns = list(county_debt.columns.values)

#Creating separate data frames for each category, and storing them into Postgresql tables.

with engine.connect() as con:
    con.execute("DROP SCHEMA IF EXISTS public CASCADE")
    con.execute('CREATE SCHEMA public')

#County data load.
n=0
print("BEGIN COUNTY DATA LOAD.")
for x in range(2, 24, 3):
        n += 1
        y = x+1
        z = y+1
        str(n)
        data_n = pd.DataFrame({'county' : dfc['County'], 
                                'state' : dfc['State'],
                            'overall': dfc.iloc[:,x ],
                            'white_areas': dfc.iloc[:,y],
                            'nonwhite_areas': dfc.iloc[:,z]})
        
        table_name = 'co_'+ KateColumns[x].replace(" ", "_").lower()
        
        data_n.to_sql(name=table_name[:50], con=engine, if_exists='append', index=False)
        
        print('')
        print(str(KateColumns[x]) + ":")
        print('')
        print(data_n.head())
        print('_____________________________________________________________ ')

print("END COUNTY DATA LOAD.")
print("--------------------------------------------")
print("BEGIN STATE DATA LOAD.")
#State data load.
KateColumns = list(state_debt.columns.values)
for x in range(1, 24, 3):
        n += 1
        y = x+1
        z = y+1
        str(n)
        data_n = pd.DataFrame({'state' : dfs['State'],
                            'overall': dfs.iloc[:,x ],
                            'white_areas': dfs.iloc[:,y],
                            'nonwhite_areas': dfs.iloc[:,z]})
        
        table_name = 'st_' + KateColumns[x].replace(" ", "_").lower()
        
        data_n.to_sql(name=table_name[:50], con=engine, if_exists='append', index=False)
        
        print('')
        print(str(KateColumns[x]) + ":")
        print('')
        print(data_n.head())
        print('_____________________________________________________________ ')

print("END STATE DATA LOAD.")
```
